####11. Loops

- Loops are used when a part of a program needs to be executed **repeatedly**.
- Many real-world applications (like displaying products in apps) use loops.
- Data is often stored in a container and fetched from database **repeatedly** using loops.
- Python has **two types of loops**:
  - `while` loop
  - `for` loop


#### 12. While Loop

- The `while` loop executes as long as the given condition is `True`.
- It checks the condition **before** each iteration.

#### Important Notes
- Always update the loop variable inside the loop.
- If the condition never becomes `False`, it will create an **infinite loop**.


```python
# Syntax

while condition:
    # code to execute
```

```python
 number = int(input("Enter the number: "))
 i = 1
 while i < 11:
  print(number, "*",i,"=", number * i)
  i = i+1
```

##### Real-Life Use of `while` Loop

- A `while` loop is used when the **number of iterations is not known in advance**.
- Execution continues **as long as the condition is true**.

Common use-cases:
- Login and authentication systems
- Games (run until win/lose condition)
- Menu-driven programs
- Reading data until it ends
- Servers and background services
- Sensor-based systems (IoT, robotics)


#### 13. Guessing Game

This is a simple number guessing game where:
- The computer generates a random number.
- The user keeps guessing until the correct number is found.
- The program gives hints after each guess.


##### Approach 1: Using `while` with a condition

- Loop continues **until the correct number is guessed**.
- A counter is used to track the number of attempts.


```python
import random

jackpot = random.randint(1, 100)
guess = int(input("Guess the number: "))
counter = 1

while guess != jackpot:
    if guess < jackpot:
        print("Guess Higher.")
    else:
        print("Guess Lower.")

    guess = int(input("Guess the number: "))
    counter += 1

print("Correct Answer.")
print("You took", counter, "attempts.")
```

##### Approach 2: Using `while True` and `break`

- `while True` creates an infinite loop.
- Loop exits using `break` when the correct guess is made.
- Attempts start from 0 because guessing happens inside the loop.


```python
import random

jackpot = random.randint(1, 100)
counter = 0

while True:
    guess = int(input("Guess the number: "))
    counter += 1

    if guess < jackpot:
        print("Guess Higher.")
    elif guess > jackpot:
        print("Guess Lower.")
    else:
        print("Correct Answer.")
        print("You took", counter, "attempts.")
        break
```

##### Approach 3: Using `for` loop (Limited Attempts)

- User gets **only 5 attempts**.
- `else` block executes if the loop finishes **without a correct guess**.


```python
import random

jackpot = random.randint(1, 100)

for attempts in range(1, 6):
    guess = int(input("Guess the number: "))

    if guess < jackpot:
        print("Guess Higher.")
    elif guess > jackpot:
        print("Guess Lower.")
    else:
        print("Correct Answer.")
        print("You took", attempts, "attempts.")
        break
else:
    print("Game Over! The number was", jackpot)
```

#####Conclusion
Approach 2 using while True and break is the most flexible and commonly used in real-world applications, while the others are useful depending on constraints like known attempts

#### 14. For Loop

- A `for` loop is used to iterate over a **sequence** or a **range**.
- It is preferred when we **know beforehand** how many times the loop will run.
- Unlike `while`, a `for` loop does **not need a termination condition**.


##### Range Function

- `range()` is used to generate a sequence of integers.
- Syntax: `range(start, stop, step)`

Examples:
- `range(1, 11)` → 1 to 10
- `range(5)` → 0 to 4 (start defaults to 0)
- `range(1, 11, 2)` → odd numbers
- `step` can also be negative


```python
print(list(range(1, 11)))
print(list(range(5)))
print(list(range(1, 11, 2)))
```

```python
for i in range(1, 11, 2):
    print(i)
```

##### Sequence

- A sequence is any ordered collection of elements.
- Examples of sequences: String, List, Tuple, Set, Dictionary.
- A `for` loop can iterate over any sequence.


```python
# Iterating over a String
for i in "kolkata":
    print(i)
```

```python
# Iterating over a List
for i in [1, 2, 3, 4]:
    print(i)
```

```python
# Iterating over a Tuple
for i in (10, 20, 30):
    print(i)
```

```python
# Iterating over a Set
for i in {1, 2, 3, 4}:
    print(i)
```

```python
student = {"name": "Radhika", "age": 20, "city": "Delhi"}

# Iterating over a Dictionary (Keys by Default)
for key in student:
    print(key)
print()

#Iterating over Dictionary Values
for value in student.values():
    print(value)
print()

# Iterating over Dictionary Key–Value Pairs
for key, value in student.items():
    print(key, ":", value)
```

##### When to Use Which Loop?

- Use `for` loop → when the number of iterations is known.
- Use `while` loop → when the number of iterations is unknown.


#### 15. Nested Loop

- A **nested loop** means a loop inside another loop.
- The inner loop runs **completely** for each iteration of the outer loop.
- Nested loops are commonly used for:
  - Pattern printing
  - Working with matrices
  - Comparing elements in collections


##### Time Complexity (Basic Idea)

- One loop → `O(n)`
- Two nested loops → `O(n²)`
- Three nested loops → `O(n³)`

Example use-cases:
- Social media: Numbers of friends of each user on social media platform.
  - Outer loop → users  
  - Inner loop → friends of each user
- Banking systems: Number of transactions of each user on a banking platform.
  - Outer loop → users  
  - Inner loop → transactions of each user


```python
# Star Pattern Using Nested Loop

rows = int(input("Enter the number of rows: "))

for i in range(1, rows + 1):
    for j in range(0, i):
        print("*", end=" ")
    print()
```

##### How the Pattern Works
- Outer loop controls the **number of rows**.
- Inner loop controls the **number of stars in each row**.
- `end=" "` keeps stars on the same line.
- `print()` moves to the next line after each row.

Use **pythontutor.com** to visualize nested loops step by step and understand how variables change during execution.


#### 16. Break, Continue and Pass

These statements are used to control the flow of loops.


##### Break

- Terminates the loop immediately when a condition is met.
- Useful when further iterations are unnecessary.
- Common use-case: **Linear search** (reduces run-time).


```python
for i in range(1, 11):
    if i == 5:
        break
    print(i)
```

##### Continue

- Skips the current iteration when a condition is met.
- The loop continues with the next iteration.
- Common use-case: Skipping unavailable/out-of-stock items.


```python
for i in range(1, 11):
    if i == 5:
        continue
    print(i)
```

##### Pass

- Acts as a **placeholder** where a statement is syntactically required.
- Does nothing: no output, no error.
- Commonly used in:
  - Loops
  - Functions
  - Classes


```python
for i in range(1, 11):
    pass
```

##### Real-Life Use of `break`, `continue`, and `pass`

- **`break`**
  - Used to **stop a loop early** when the goal is achieved.
  - Examples:
    - Searching for an item and stopping once it is found

- **`continue`**
  - Used to **skip specific cases** and move to the next iteration.
  - Examples:
    - Skipping out-of-stock products in an e-commerce app
    - Ignoring invalid or missing data while processing records

- **`pass`**
  - Used as a **placeholder** when logic is not yet implemented.
  - Examples:
    - Creating empty functions or classes during planning
    - Writing code structure first and filling logic later
