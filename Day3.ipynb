#### **20. Accessing Substrings from a String**

- Strings are indexed sequences.
- Python supports:
  - Positive indexing (left -> right, starts from 0)
  - Negative indexing (right -> left, starts from -1)
- Extracting multiple characters from a string is called **slicing**.
- Slicing syntax: `string[start : end : step]`


```python
# Indexing

c = "Hello"
print(c)

# Positive indexing
print(c[3])    # 'l'

# Negative indexing
print(c[-5])   # 'H'
```

```python
# Slicing

c = "Hello World"

print(c[0:5])    # 'Hello'
print(c[2:])     # From index 2 till end
print(c[:4])     # From start till index 3
print(c[:])      # Entire string
print(c[2:6:2])  # start, end, step
```

```python
# Using 'negative step' when 'start' is smaller than 'end' gives empty string
print(c[0:6:-1])

# Negative index slicing
print(c[-5:-1:2])

# Reverse string
print(c[::-1])

# Reverse using negative indices
print(c[-1:-5:-1])
```

##### Key Points
- `end` index is always exclusive
- Positive indexing cannot work with negative step
- `[::-1]` is the simplest way to reverse a string


#### **21. Editing and Deleting Strings**

- Strings are **immutable** in Python.
- Individual characters of a string **cannot be modified**.
- Editing a string is not possible, but the variable can be **reassigned**.
- Individual characters cannot be deleted.
- Entire string objects can be deleted using `del`.


```python
c = "Hello"
print(c)

# c[0] = 'X'   # TypeError (strings are immutable)

# Reassigning the string
c = "world"
print(c)

# c[5] = 'X'   # Not allowed

# del c[0]     # Not allowed

# Deleting the entire string
del c
# print(c)     # NameError: c is not defined
```

#### **22. Operations on Strings**

Strings support multiple types of operations:
- Arithmetic operations
- Relational operations
- Logical operations
- Loops on strings
- Membership operations


```python
# Arithmetic operations
print("Hello" + " World")   # Concatenation
print("*" * 10)             # Repetition
```

```python
# Relational operations (lexicographical comparison)
print("Hello" == "World")
print("Hello" != "World")
print("Mumbai" > "Pune")    # Dictionary order, comparison is done lexiographically
print("Goa" < "Kolkata")
print("goa" < "Goa")        # Lowercase comes after uppercase
```

```python
# Logical operations
# Empty string -> False, Non-empty string -> True
print("hello" and "world") # "hello" is 1, "world" is 1, "world" will be output, because after world "and" will evaluate results.
print("hello" or "world") # "hello" is 1, "world" is 1, "hello" will be output, because after hello "or" can evaluate results.
print("" or "world") # "" is 0, "world" is 1, "world" will be output.
print("" and "world") # "" is 0, "world" is 1, "" will be output, because 1 and 0 is 0.
print(not "") # "" is 0, so output will be true.
print(not "Hello") # output will be false.
```

```python
# Loops on strings
c = "Hello World"
for i in c:
    print(i)

print()

# Loop with slicing
for i in c[0:12:2]:
    print(i)
```

```python
# Membership operations
print("H" in c)
print("World" not in c)
```

#### **23. String Functions**

Python provides many built-in string functions used for:
- Length and ordering
- Case conversion
- Searching and counting
- Validation
- Splitting and joining
- Replacing and trimming strings


```python
# Common Functions (Iterables)

c = "kolkata"

print(len(c))
print(max(c))                 # Based on ASCII values
print(min(c))                 # Based on ASCII values
print(sorted(c))              # Ascending order (list output)
print(sorted(c, reverse=True)) # Descending order
```

```python
# Case Conversion Functions

c = "it is raining today."

print(c.capitalize())   # First letter capital
print(c.title())        # First letter of each word capital
print(c.upper())        # All uppercase
print(c.lower())        # All lowercase
print(c.swapcase())     # Swap upper ↔ lower
```

```python
# Count Function

print(c.count("is"))
print(c.count("x"))     # Returns 0 if not found
```

```python
# Find vs Index

print(c.find("i"))        # First occurrence
print(c.find("raining"))
print(c.find("x"))        # Returns -1 if not found

print(c.index("raining"))
# print(c.index("x"))     # Error if not found
```

```python
# startswith / endswith

print(c.endswith("raining."))
print(c.startswith("i"))
```

```python
# format()

print("Hello my name is {} and I am {}.".format("Radhika", 21))
print("Hello my name is {1} and I am {0}.".format("Radhika", 21))
print("Hello my name is {name} and I am {age}.".format(name="Radhika", age=21))
```

```python
# Validation Functions

d = "flat20"

print(d.isalnum())
print(d.isalpha())
print(d.isascii())
print("20".isdigit())
print("helloworld".isidentifier())
```

```python
# split()

print(c.split())
print(c.split("is"))
```

```python
# join()

print(" ".join(["My", "name", "is", "Radhika"]))
print("-".join(["My", "name", "is", "Radhika"]))
```

```python
# replace()

print("My name is Radhika".replace("Radhika", "Utkarsh"))
```

```python
# strip()

name = "     radhika   "
print(name.strip())
```

##### Key Points
- String functions return **new strings** (strings are immutable)
- `find()` returns `-1`, `index()` throws error if not found
- `split()` breaks string into list, `join()` combines list into string
- Validation functions are useful for input checking
